# Image Processing

```python
from scipy import misc
import matplotlib.pyplot as plt
ascent = misc.ascent()
plt.axis("off")
plt.gray()
plt.imshow(ascent)
#We can get rid of the ticks and the axis by adding the command plt.axis("off") :
plt.show()
```

```python
import scipy.misc
face = scipy.misc.face()
print(face.shape)
print(face.max)
print(face.dtype)
plt.axis("off")
plt.gray()
plt.imshow(face)
plt.show()
```

## Tilt, Shade, Tone

```python
import numpy as np
import matplotlib.pyplot as plt

def tint(image, percent):
    tinted_image = image + (np.ones(image.shape)- image) * percent
    return tinted_image

football = plt.imread('b7.png')
tinted_football = tint(football, 0.6)
plt.axis("off")
plt.imshow(tinted_football)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tint(image, percent):
    shade_image = image - (1 *  percent)
    return shade_image

football = plt.imread('b7.png')
shade_football = tint(football, 0.6)
plt.axis("off")
plt.imshow(shade_football)
```
