#### Stock Market Performance Analysis using Python

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
startDate = datetime.now() - pd. DateOffset(months=3)
endDate = datetime.now()

tickers = ['AAPL', 'MSFT', 'NFLX', 'GOOG', 'WMT']
df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=startDate, end = endDate)
    df_list.append(data)

df = pd.concat(df_list, keys= tickers, names= ['Ticker', 'Date'])
print(df.head())
```

```python
## RESETTING THE INDEX COLUMN
df = df.reset_index()
print(df.head())
```

### performance in the stock market of all the companies:

```python

import plotly.express as px
fig = px.line( df, 
              x= "Date",
              y = 'Close',
              color= 'Ticker',
              title = 'Stock Market Performance for the Last 3 months' )
fig.show()
```

### Faceted area Chart

```python
fig = px.area(df, x= 'Date', y = 'Close',
 color= "Ticker", 
facet_col= 'Ticker',
labels ={"Date": "Date", "Close": "Closing Price", "Ticker":"Company"},
 title='Stock Prices for Apple, Microsoft, Netflix, Walmart, and Google')


fig.show()
```
