# Understanding Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph 

```python
x = ["A","B", "C", "D", "E", "F", "G"]
y = [0, 2, 4, 6,8, 10, 12]
plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Days - Temp Distribution")

plt.legend("view")
plt.show()
```
